```python
#import libraries
import pandas as pd
import numpy as np
#load datasets
cases = pd.read_csv("covid_19_india.csv")
testing = pd.read_csv("StatewiseTestingDetails.csv")
vaccine = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
#shape 
print("Cases:", cases.shape)
print("Testing:", testing.shape)
print("Vaccine:", vaccine.shape)
```

```python
#standardize column name as state which in cases is state/unionterritory
#renaming colummn -cases.rename(columns={'State/UnionTerritory':'State'}, inplace=True)

cases.rename(columns={
    'State/UnionTerritory': 'State',
    'Date': 'Date'
}, inplace=True)
```

```python
#check missing values 
testing.isna().sum()
```

```python
#check missing values 
vaccine.isna().sum()
```

```python
#check missing values 
cases.duplicated().sum()
```

```python
#remove duplicates
cases.drop_duplicates(inplace=True)
testing.drop_duplicates(inplace=True)
vaccine.drop_duplicates(inplace=True)
```

```python
cases.duplicated().sum()
```

```python
#fill missing values 
cases[['Confirmed','Deaths','Cured']] = cases[['Confirmed','Deaths','Cured']].fillna(0)
testing[['TotalSamples','Negative','Positive']] = testing[['TotalSamples','Negative','Positive']].fillna(0)
numeric_cols = vaccine.select_dtypes(include='number').columns
vaccine[numeric_cols] = vaccine[numeric_cols].fillna(0)
```

```python
#date column cleaning 
cases['Date'] = pd.to_datetime(cases['Date'], errors='coerce')
testing['Date'] = pd.to_datetime(testing['Date'], errors='coerce')
vaccine['Updated On'] = pd.to_datetime(vaccine['Updated On'], errors='coerce')
```

```python
#recovery and death rate 
cases['Recovery_Rate'] = cases['Cured'] / cases['Confirmed']
cases['Death_Rate'] = cases['Deaths'] / cases['Confirmed']
```

```python
#month wise total confirmed cases 
cases.groupby(cases['Date'].dt.month)['Confirmed'].sum()
```

```python
#state wise total confirmed cases
cases.groupby('State/UnionTerritory')['Confirmed'].sum().sort_values(ascending=False)
```

```python
#why positivity rate matters 
#testing['Positivity_Rate'] = testing['Positive'] / testing['TotalSamples']
testing['Positivity_Rate'] = np.where(
    testing['TotalSamples'] > 0,
    testing['Positive'] / testing['TotalSamples'],
    0
)
```

```python
#fully vaccinated = 2nd dose completed
vaccine['Fully_Vaccinated'] = vaccine['Second Dose Administered']
#gender coverage
vaccine['Gender_Coverage'] = vaccine['Male Doses'] + vaccine['Female Doses'] + vaccine['Transgender Doses']
```

```python
#recovery or death rate 
cases['Recovery_Rate'] = np.where(
    cases['Confirmed'] > 0,
    cases['Cured'] / cases['Confirmed'],
    0
)

cases['Death_Rate'] = np.where(
    cases['Confirmed'] > 0,
    cases['Deaths'] / cases['Confirmed'],
    0
)
```

```python
#time based analysis 
cases = cases.sort_values('Date')
```

```python
#daily confirmed cases 
daily_cases = cases.groupby('Date')['Confirmed'].sum().reset_index()
daily_cases.head()
```

```python
#monthly confirmed cases 
monthly_cases = (
    cases
    .groupby(cases['Date'].dt.to_period('M'))['Confirmed']
    .sum()
    .reset_index()
)

monthly_cases['Date'] = monthly_cases['Date'].astype(str)
monthly_cases.head()
```

```python
daily_cases.to_csv("daily_confirmed_cases.csv", index=False)
monthly_cases.to_csv("monthly_confirmed_cases.csv", index=False)
```

```python
#state wise analysis 
#total cases by state 
state_cases = (
    cases
    .groupby('State')['Confirmed']
    .sum()
    .reset_index()
    .sort_values(by='Confirmed', ascending=False)
)

state_cases.head(5)
```

```python
#state wise death rate 
state_death_rate = (
    cases
    .groupby('State')
    .agg({
        'Deaths':'sum',
        'Confirmed':'sum'
    })
    .reset_index()
)

state_death_rate['Death_Rate'] = (
    state_death_rate['Deaths'] /
    state_death_rate['Confirmed']
)

state_death_rate.sort_values('Death_Rate', ascending=False).head()
```

```python
state_cases.to_csv("state_wise_total_cases.csv", index=False)
state_death_rate.to_csv("state_wise_death_rate.csv", index=False)
```

```python
#merging 
vaccine.rename(columns={'Updated On': 'Date'}, inplace=True)
```

```python
#merge cases and testing
cases_testing = pd.merge(
    cases,
    testing,
    on=['State', 'Date'],
    how='left'
)

cases_testing.head()
```

```python
#merged with vaccine data
full_data = pd.merge(
    cases_testing,
    vaccine,
    on=['State', 'Date'],
    how='left'
)

full_data.head()
```

```python
full_data.to_csv("covid_full_merged_data.csv", index=False)
```

```python
#vaccination impact vs cases 
#mmonthly vaccination vs cases 
monthly_vaccine_cases = (
    full_data
    .groupby(full_data['Date'].dt.to_period('M'))
    .agg({
        'Confirmed': 'sum',
        'Second Dose Administered': 'sum'
    })
    .reset_index()
)

monthly_vaccine_cases['Date'] = monthly_vaccine_cases['Date'].astype(str)
monthly_vaccine_cases.head()
#save vaccination vs cases csv file
monthly_vaccine_cases.to_csv(
    "monthly_vaccination_vs_cases.csv",
    index=False
)
```

```python
#positivity rate
#statewise positivity rate 
state_positivity = (
    testing
    .groupby('State')
    .agg({
        'Positive': 'sum',
        'TotalSamples': 'sum'
    })
    .reset_index()
)

state_positivity['Positivity_Rate'] = (
    state_positivity['Positive'] /
    state_positivity['TotalSamples']
)

state_positivity.sort_values(
    'Positivity_Rate',
    ascending=False
).head()
```

```python
#save positivity report 
state_positivity.to_csv(
    "state_wise_positivity_rate.csv",
    index=False
)
```

```python
#advanced level
#active cases 
cases['Active_Cases'] = (
    cases['Confirmed'] -
    cases['Cured'] -
    cases['Deaths']
)
```

```python
#daily new cases - identifies spikes 
cases = cases.sort_values(['State', 'Date'])

cases['Daily_New_Cases'] = (
    cases.groupby('State')['Confirmed']
    .diff()
    .fillna(0)
)
```

```python
#7 day rolling average 
cases['7Day_Avg_New_Cases'] = (
    cases.groupby('State')['Daily_New_Cases']
    .rolling(7)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
#case fatality rate
cases['Case_Fatality_Rate'] = np.where(
    cases['Confirmed'] > 0,
    cases['Deaths'] / cases['Confirmed'],
    0
)
```

```python
#vaccination coverage 
vaccine['Vaccination_Coverage'] = np.where(
    vaccine['Total Doses Administered'] > 0,
    vaccine['Second Dose Administered'] /
    vaccine['Total Doses Administered'],
    0
)
```

```python
#testing intensity 
testing['Tests_Per_Positive'] = np.where(
    testing['Positive'] > 0,
    testing['TotalSamples'] / testing['Positive'],
    0
)
```

```python
#growth rate(month over month )
monthly_cases['MoM_Growth_Rate'] = (
    monthly_cases['Confirmed']
    .pct_change()
    .fillna(0)
)
```
